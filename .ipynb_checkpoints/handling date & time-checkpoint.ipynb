```python
df1['date']=pd.to_datetime(df1['date'])
```

```python
df1.info()
```

```python
df1=df1.drop(columns=['product_id','city_id','orders'])
```

```python
df1.head()
```

```python
df1['date_year']=df1['date'].dt.year
```

```python
df1.head()
```

```python
df1['date_month']=df1['date'].dt.month
#for month name
df1['month name']=df1['date'].dt.month_name()
```

```python
df1.head()
```

```python
df1['date_day']=df1['date'].dt.day
#for day name
df1['date_day_name']=df1['date'].dt.day_name()
```

```python
df1.head()
```

```python
df1['date_day_of_week']=df1['date'].dt.dayofweek
```

```python
df1.head()
```

```python
df1['weekend or not']=np.where(df1['date_day_name'].isin(['Sunday','Monday']),1,0)
```

```python
df1.head()
```

```python
df1['date_week'] = df1['date'].dt.isocalendar().week
```

```python
df1.head()
```

```python
df1['quarter'] = df1['date'].dt.quarter
```

```python
df1.head()
```

```python
df1['semester']=np.where(df1['quarter'].isin([1,2]),1,2)
```

```python
df1.head()
```
